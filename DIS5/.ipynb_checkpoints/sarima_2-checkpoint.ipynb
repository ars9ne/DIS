```python
import pandas as pd
import numpy as np
from pmdarima import auto_arima
import matplotlib.pyplot as plt

# Загрузка и предобработка данных
data = pd.read_csv('Bitcoin Historical Data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Price'] = data['Price'].str.replace(',', '').astype(float)
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
price_data = data['Price']

# Применение логарифмической трансформации
price_data_log = np.log(price_data)

# Использование данных до конца 2022 года
price_data_train_log = price_data_log[:'2022']

# Автоматический подбор параметров SARIMA
sarima_model = auto_arima(price_data_train_log, 
                          start_p=0, start_q=0, 
                          max_p=3, max_q=3,
                          seasonal=True,
                          m=48,  # сезонный период 48 месяцев (4 года)
                          start_P=0, start_Q=0, 
                          max_P=2, max_Q=2,
                          d=1, D=1,  # порядок дифференцирования
                          trace=True,
                          error_action='ignore',
                          suppress_warnings=True,
                          stepwise=True)

# Вывод выбранных параметров
print(sarima_model.summary())
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Загрузка и предобработка данных
data = pd.read_csv('Bitcoin Historical Data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Price'] = data['Price'].str.replace(',', '').astype(float)
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
price_data = data['Price']

# Логарифмическое преобразование цен
price_data_log = np.log(price_data)

# Используем данные до конца 2022 года
price_data_train_log = price_data_log[:'2022']

# Построение и обучение модели SARIMA с новыми параметрами
model = SARIMAX(price_data_train_log,
                order=(2, 1, 2),                # (p, d, q) - измененные параметры
                seasonal_order=(1, 1, 1, 48),   # (P, D, Q, m) - сезонные компоненты
                enforce_stationarity=False,
                enforce_invertibility=False)

fitted_model = model.fit(disp=False)
print(fitted_model.summary())

```

```python

# Прогноз на следующие 4 года (48 месяцев)
forecast = fitted_model.get_forecast(steps=192)
forecast_index = pd.date_range(start=price_data_train_log.index[-1], periods=193, freq='M')[1:]
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)

# Обратное преобразование прогноза в оригинальную шкалу
forecast_series_original_scale = np.exp(forecast_series)

# Построение графика результатов
plt.figure(figsize=(10, 6))
plt.plot(price_data, label='Исторические данные по ценам')
plt.plot(forecast_series_original_scale, label='Прогнозируемые цены (оригинальная шкала)')
plt.title('Прогноз цен на биткоин')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.show()
```
