```python
import pandas as pd
import numpy as np
from pygam import LinearGAM, s
import matplotlib.pyplot as plt

# Загрузка и предобработка данных
data = pd.read_csv('Bitcoin Historical Data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Price'] = data['Price'].str.replace(',', '').astype(float)
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
price_data = data['Price']

# Преобразование данных в формат для GAM
df = price_data.reset_index()
df['Time'] = np.arange(len(df))

# Создание и обучение модели GAM с одной переменной
gam = LinearGAM(s(0)).fit(df[['Time']], np.log(df['Price']))
```

```python

# Прогнозирование
future = pd.DataFrame({'Time': np.arange(len(df), len(df) + 48)})
forecast = gam.predict(future)

# Преобразование прогноза обратно в оригинальную шкалу
forecast_original_scale = np.exp(forecast)

# Построение графика результатов
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'], label='Историческая цена')
plt.plot(pd.date_range(start=df['Date'].iloc[-1], periods=49, freq='M')[1:], forecast_original_scale, label='Прогнозируемые цены')
plt.title('Прогноз цен на BTC с помощью GAM')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from pygam import LinearGAM, s, te
import matplotlib.pyplot as plt
data = pd.read_csv('Bitcoin Historical Data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Price'] = data['Price'].str.replace(',', '').astype(float)
data['High'] = data['High'].str.replace(',', '').astype(float)
data['Low'] = data['Low'].str.replace(',', '').astype(float)
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
data['Price_Change'] = data['Price'].pct_change().fillna(0)
data['Day_of_Week'] = data.index.dayofweek
data['Month'] = data.index.month
train_data = data[:'2022-12-31']
test_data = data['2023-01-01':]
train_df = train_data.reset_index()
train_df['Time'] = np.arange(len(train_df))
X_train = train_df[['Time', 'Price_Change', 'High', 'Low', 'Day_of_Week', 'Month']]
y_train = np.log(train_df['Price'])
gam = LinearGAM(te(0, 1) + s(2) + s(3) + s(4) + s(5)).fit(X_train, y_train)

```

```python
test_df = test_data.reset_index()
test_df['Time'] = np.arange(len(train_df), len(train_df) + len(test_df))
X_test = test_df[['Time', 'Price_Change', 'High', 'Low', 'Day_of_Week', 'Month']]
forecast_log = gam.predict(X_test)
forecast_original_scale = np.exp(forecast_log)
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Price'], label='Историческая цена')
plt.plot(test_df['Date'], forecast_original_scale, label='Прогнозируемые цены', linestyle='--')
plt.axvline(x=pd.Timestamp('2023-01-01'), color='r', linestyle='--', label='Начало прогноза')
plt.title('Прогноз цен на BTC с помощью модели GAM с 2023 года')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from pygam import LinearGAM, s, te
import matplotlib.pyplot as plt

# Загрузка и предобработка данных
data = pd.read_csv('Bitcoin Historical Data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Price'] = data['Price'].str.replace(',', '').astype(float)
data['High'] = data['High'].str.replace(',', '').astype(float)
data['Low'] = data['Low'].str.replace(',', '').astype(float)
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
data['Price_Change'] = data['Price'].pct_change().fillna(0)
data['Day_of_Week'] = data.index.dayofweek
data['Month'] = data.index.month

# Разделение данных на обучающую и тестовую выборки
train_data = data[:'2022-12-31']
test_data = data['2023-01-01':]

# Преобразование данных в формат для GAM
train_df = train_data.reset_index()
train_df['Time'] = np.arange(len(train_df))

# Создание и обучение модели GAM с дополнительными временными признаками
X_train = train_df[['Time', 'Price_Change', 'High', 'Low', 'Day_of_Week', 'Month']]
y_train = np.log(train_df['Price'])

gam = LinearGAM(te(0, 1) + s(2) + s(3) + s(4) + s(5)).fit(X_train, y_train)

# Прогнозирование на тестовом периоде и на год вперед
future_dates = pd.date_range(start=test_data.index[-1] + pd.Timedelta(days=1), periods=365)
future_df = pd.DataFrame(index=future_dates)
future_df['High'] = test_data['High'].mean()  # Предположим, что High будет на среднем уровне
future_df['Low'] = test_data['Low'].mean()  # Предположим, что Low будет на среднем уровне
future_df['Price_Change'] = test_data['Price_Change'].mean()  # Предположим, что Price Change будет на среднем уровне
future_df['Day_of_Week'] = future_df.index.dayofweek
future_df['Month'] = future_df.index.month
future_df['Time'] = np.arange(len(train_df) + len(test_data), len(train_df) + len(test_data) + len(future_df))

# Объединение тестового периода и будущих данных
test_and_future_df = pd.concat([test_data.reset_index(), future_df.reset_index()])

# Замена значений NaN и Inf на средние значения
test_and_future_df = test_and_future_df.replace([np.inf, -np.inf], np.nan)
test_and_future_df = test_and_future_df.fillna(test_and_future_df.mean())

# Проверка размера данных
print(test_and_future_df.shape)

# Прогнозирование
X_test_and_future = test_and_future_df[['Time', 'Price_Change', 'High', 'Low', 'Day_of_Week', 'Month']]
forecast_log = gam.predict(X_test_and_future)

# Преобразование прогноза обратно в оригинальную шкалу
forecast_original_scale = np.exp(forecast_log)

# Построение графика результатов
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Price'], label='Историческая цена')
plt.plot(test_and_future_df['Date'], forecast_original_scale, label='Прогнозируемые цены', linestyle='--')
plt.axvline(x=pd.Timestamp('2023-01-01'), color='r', linestyle='--', label='Начало прогноза')
plt.title('Прогноз цен на BTC с помощью модели GAM с 2023 года (без объема)')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.show()
```
