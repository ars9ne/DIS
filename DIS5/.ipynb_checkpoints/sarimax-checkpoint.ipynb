```python
import pandas as pd

# Load the data
file_path = 'Bitcoin Historical Data.csv'
bitcoin_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
bitcoin_data.head()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt


# Load and preprocess the data
data = pd.read_csv('Bitcoin Historical Data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data['Price'] = data['Price'].str.replace(',', '').astype(float)
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
price_data = data['Price']

# Apply logarithmic transformation
price_data_log = np.log(price_data)

# Use data up to the end of 2022
price_data_train_log = price_data_log[:'2022']

# Check for stationarity
result = adfuller(price_data_train_log.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Fit the SARIMA model
model = SARIMAX(price_data_train_log,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 48),  # Seasonal period as 4 years (48 months)
                enforce_stationarity=False,
                enforce_invertibility=False)

fitted_model = model.fit(disp=False)
print(fitted_model.summary())
```

```python
# Forecast for the next 4 years (48 months)
forecast = fitted_model.get_forecast(steps=192)
forecast_index = pd.date_range(start=price_data_train_log.index[-1], periods=193, freq='M')[1:]
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)

# Convert forecast back to original scale (exponential)
forecast_series_original_scale = np.exp(forecast_series)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(price_data, label='Historical Monthly Prices')
plt.plot(forecast_series_original_scale, label='Forecasted Prices (Original Scale)')
plt.title('Bitcoin Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
